# VISUALIAZATION

```python
plt.figure(figsize=(10,6))
plt.title('Number of confirmed cases by State in India')
sns.barplot(x=covid_cases_India_df['Name of State / UT'],y=covid_cases_India_df['Total Confirmed cases'])
plt.ylabel('Number of Cases')
# Rotating xlabels so that we can read the text displayed
plt.xticks(rotation=45,horizontalalignment='right',fontweight='light')
```

# Exact cases in Maharashtra?

```python
covid_cases_India_df[covid_cases_India_df['Name of State / UT']=='Maharshtra']
```

# Number of Cured/Discharged/Migrated cases by state

```python
plt.figure(figsize=(10,6))
plt.title(' Number of Cured/Discharged/Migrated cases by state')
sns.barplot(x=covid_cases_India_df['Name of State / UT'],y=covid_cases_India_df['Cured/Discharged/Migrated'])
plt.ylabel('Number of cases')
plt.xticks(rotation='45',horizontalalignment='right',fontweight='light')
```

# Number of Deaths by state

```python
plt.figure(figsize=(10,6))
plt.title('Number of deaths by State')
sns.barplot(x=covid_cases_India_df['Name of State / UT'],y=covid_cases_India_df['Deaths'])
plt.ylabel('Number of Cases')
plt.xticks(
    rotation=45, 
    horizontalalignment='right',
    fontweight='light' 
)
```
